# Загрузка полного содержания статей (PDF)

Временное ограничение на 3000 .pdf файлов научных статей (для предварительного анализа и обработки). В случае необходимости данное ограничение можно убрать.

```python
import os
import pandas as pd
from doi2pdf import doi2pdf
from tqdm import tqdm

os.makedirs("pdfs", exist_ok=True)

df = pd.read_csv("cleaned_dataset.csv")

dois = df['doi'].dropna()
count = 0
for doi in tqdm(dois, desc="Загрузка arXiv PDF"):
    try:
        if count == 3000:
          break
        filename = doi.replace("/", "_") + ".pdf"
        output_path = os.path.join("pdfs", filename)
        doi2pdf(doi, output=output_path)
        count += 1
    except Exception as e:
        print(f"Ошибка при загрузке {doi}: {e}")
```

# Основная часть

```python
import os
import pandas as pd
import numpy as np
import ast
from tqdm import tqdm
from PyPDF2 import PdfReader
from sentence_transformers import SentenceTransformer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

## Загружаем необходимую модель

В качестве модели можно взять небольшую модель с Hugging  Face, например, "sentence-transformers/all-MiniLM-L6-v2". Также можно обратить внимание на более крупные модели или модель "allenai-specter", которая была предобучена на научных статьях (ей и воспользуемся).

```python
model = SentenceTransformer("allenai-specter")
```

Начинаем строить таблицу df с данными, которые потом будут использоваться для построения эмбеддингов.

```python
df = pd.read_csv("cleaned_dataset.csv")
df['abstract'] = df['abstract'].fillna("")
df['title'] = df['title'].fillna("")
df['doc_type'] = df['doc_type'].fillna("")
df['doi'] = df['doi'].astype(str).replace('nan', '')

def extract_fos_names(fos_json):
    try:
        fos_list = ast.literal_eval(fos_json)
        return " ".join(d["name"] for d in fos_list)
    except:
        return ""

df['fos_str'] = df['fos'].apply(extract_fos_names)


def extract_pdf_text(doi):
    if not doi or doi == 'nan':
        return ""
    try:
        pdf_path = os.path.join("pdfs", doi.replace("/", "_") + ".pdf")
        if os.path.exists(pdf_path):
            try:
                reader = PdfReader(pdf_path)
                return " ".join([page.extract_text() or "" for page in reader.pages])
            except:
                return ""
    except AttributeError:
        return ""
    return ""

df['pdf_text'] = df['doi'].apply(extract_pdf_text)
df['pdf_text'] = df['pdf_text'].fillna("")
```

```python
example_text = df[df['pdf_text'].str.strip() != ""].iloc[0]['pdf_text']
print(example_text[:2000])
```

Смотрим, что самые важные поля ('title', 'abstract', 'pdf_text') не пустые. Создаем тексты, которые будем передавать для построение эмбеддингов нашей модели model.

```python
df = df[
    (df['title'].str.strip() != "") &
    (df['abstract'].str.strip() != "") &
    (df['pdf_text'].str.strip() != "")
].copy()


df['text'] = (
    df['title'] + ". " +
    df['abstract'] + ". " +
    df['pdf_text']
)


if 'doc_type' in df.columns:
    df['text'] += ". " + df['doc_type'].fillna("")

if 'fos_str' in df.columns:
    df['text'] += ". " + df['fos_str'].fillna("")


texts = df['text']
```

Сохраняем результат после построение эмбеддингов и разбиваем нашу выборку на train и test в отношениях 80% к 20% соответственно.

```python
embed_path = "trained_models/embeddings_fulltext.npy"

os.makedirs("trained_models", exist_ok=True)

if os.path.exists(embed_path):
    print("Загрузка кэшированных эмбеддингов...")
    embeddings = np.load(embed_path)
else:
    print("Генерация эмбеддингов")
    embeddings = model.encode(
        texts.tolist(),
        batch_size=32,
        show_progress_bar=True,
        convert_to_numpy=True,
    )
    np.save(embed_path, embeddings)

year_values = df['year'].fillna(df['year'].median()).values.reshape(-1, 1)

X_full = np.hstack([embeddings, year_values])
y_full = df['n_citation'].fillna(0).values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_full)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_full, test_size=0.2, random_state=42)
```

Наконец обучаем подходящую модель на нашей выборке. По предварительным результатам ElasticNetCV показал наилучший результат, например, по метрикам $R^2$ и $RMSE$.

```python
from sklearn.linear_model import ElasticNetCV
from sklearn.metrics import r2_score, root_mean_squared_error

regressor = ElasticNetCV(tol=1e-3, max_iter=10000)
regressor.fit(X_train, y_train)

y_pred = regressor.predict(X_test)
print("R²:", r2_score(y_test, y_pred))
print("RMSE:", root_mean_squared_error(y_test, y_pred))
```
